```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes3=fig.add_axes([0.65,0.15,0.2,0.2])
axes3.plot(x,y,color='purple',linewidth=5)
axes2.plot(x,y,'r-',linewidth=5)
axes1.plot(x,y,'b-')
axes3.set_title('Smaller Plot2')
axes2.set_title('Smaller Plot1')
axes1.set_title('Larger Plot')

#(left,bottom,width,height)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Soja')
axes[1].plot(y,x,'y')
axes[1].set_title('Ulto')
plt.tight_layout()
```

```python
#Figure Size and DPI
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
#(left,bottom,width,height)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,6))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
#(left,bottom,width,height)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.set_title('graph')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**3,label='X cube')
ax.plot(x,x**2,label='X square')
ax.set_title('graph')
ax.legend(loc=0)
```

```python
#Plot Appearance
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,alpha=1,linestyle='solid',marker='o',markersize=15,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='black')
#RGB hex code for color
#linewidth=lw
#linestyle=ls
```
