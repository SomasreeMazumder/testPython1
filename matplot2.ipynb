```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
#(left,bottom,width,height)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y,'r-')
axes1.plot(x,y,'r-')
axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')
#(left,bottom,width,height)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])

axes1.plot(x,z)
axes1.set_label('x')
axes1.set_label('y')



axes2.plot(x,y)
axes2.set_title('zoom')
axes2.set_label('x')
axes2.set_label('y')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,color='blue',ls='solid',lw=3)
axes[1].plot(x,z,color='red',lw=3,ls='--')
```
